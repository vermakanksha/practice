```python
import pandas as pd
```

## Web Scraping 250 best rated bollywood movies from IMDb website

```python
df = pd.read_html('http://www.imdb.com/india/top-rated-indian-movies/')
table = df[0]
IMDb_rating = table.iloc[:,[1,2]]
IMDb_rating.head()
```

```python
rank = IMDb_rating["Rank & Title"].str.split(pat=".",n=1,expand=True)
rank.head(3)
```

```python
movie_name = rank[1].str[:-6].str.title()
```

```python
year = rank[1].str[-5:].str.replace(")","")
```

```python
imdb_df = pd.concat([rank[0],movie_name,year,IMDb_rating["IMDb Rating"]],axis=1)
```

```python
imdb_df.columns = ["Rank","Movie_Name","Year","Rating"]
```
